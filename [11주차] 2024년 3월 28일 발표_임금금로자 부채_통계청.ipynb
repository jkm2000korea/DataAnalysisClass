### 1. 임금근로자의 개인대출 및 연체율

#### □ 2022년 12월말 기준 임금근로자의 평균대출은 5,115만원으로 전년 대비 1.7%(△87만원) 감소  
 ##### ○ 연체율(대출잔액 기준)은 0.43%로 전년보다 0.02%p 높아짐

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

# 데이터
years = ['\'17.12', '\'18.12', '\'19.12', '\'20.12', '\'21.12', '\'22.12']
avg_loan_amounts = [3974, 4175, 4408, 4862, 5202, 5115]
growth_rates = [None, 5.1, 5.6, 10.3, 7.0, -1.7]  # 첫 번째 데이터가 증감률이 없으므로 None으로 표시

# 그래프 생성
fig, ax1 = plt.subplots(figsize=(10, 6))

# 평균대출 막대그래프
ax1.bar(years, avg_loan_amounts, color='navy', label='평균대출')
# ax1.set_xlabel('연도')
ax1.set_ylabel('(만원)', color='black', rotation=0, loc='top', labelpad=(-1), fontsize=12) 
ax1.tick_params(axis='y', labelsize=12, direction='out', color='black', length=5, pad=15)

ax1.set_ylim(0, 6000)  # 첫 번째 y 축 한계 설정
ax1.tick_params(axis='x', labelsize=15)
ax1.set_yticks([0, 2000, 4000, 6000])

# 증감률 꺾은선 그래프
ax2 = ax1.twinx()
ax2.plot(years, growth_rates, marker='o', color='orange', label='증감률', markersize=10, markerfacecolor='white', markeredgewidth=2)
ax2.set_ylabel('(%)', color='black', rotation=0, loc='top', labelpad=20, fontsize=12) 
ax2.tick_params(axis='y', labelsize=12, direction='out', color='black', length=5, pad=15)

ax2.set_ylim(-5, 20)  # 두 번째 y 축 한계 설정

# 증감률 표식 설정
for i, avg in enumerate(avg_loan_amounts):
    if avg is not None:
        ax1.text(i, avg+100, f'{avg:,.0f}', ha='center', va='bottom', color='black', weight='extra bold', fontsize=14) 
        
# 증감률 표식 설정
for i, rate in enumerate(growth_rates):
    if rate is not None:
        ax2.text(i, rate+1, f'{rate}', ha='center', va='bottom', color='white', weight='extra bold', fontsize=14) 

# 그래프 제목 설정
plt.title('< 임금근로자 평균대출 >', fontsize=20, weight='extra bold')

# 범례 추가
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
lines = lines1 + lines2
labels = labels1 + labels2


# 범례 추가
plt.legend(lines, labels, loc=8, frameon=False)
# 범례 추가
plt.legend(lines, labels, bbox_to_anchor=(0.5, -0.1), loc='center', ncol=2, frameon=False)

# 그리드 제거
ax1.grid(False)
ax2.grid(False)

# 그래프 표시
plt.show()
```

```python
import matplotlib.pyplot as plt

# 데이터
years = ['\'17.12', '\'18.12', '\'19.12', '\'20.12', '\'21.12', '\'22.12']
bank = [0.23, 0.24, 0.26, 0.20, 0.16, 0.17]
non_bank = [1.13, 1.25, 1.22, 1.07, 0.87, 0.88]
total = [0.56, 0.61, 0.60, 0.50, 0.41, 0.43]

# 그래프 생성
plt.figure(figsize=(10, 6))

# 은행 데이터 꺾은선 그래프
plt.plot(years, bank, marker='o', color='skyblue', markersize=50, label='은행', markerfacecolor='skyblue', markeredgewidth=2)
for i, num in enumerate(bank):
    plt.text(i, num-0.03, f'{num}', ha='center', va='bottom', fontsize=14, color='w', weight='extra bold')

# 비은행 데이터 꺾은선 그래프
plt.plot(years, non_bank, marker='o', color='lightgrey', markersize=50, label='비은행', markerfacecolor='lightgrey', markeredgewidth=2)
for i, num in enumerate(non_bank):
    plt.text(i, num-0.03, f'{num}', ha='center', va='bottom', fontsize=14, color='w', weight='extra bold')

# 전체 데이터 꺾은선 그래프
plt.plot(years, total, marker='o', color='navy', markersize=50, label='전체', markerfacecolor='navy', markeredgewidth=2)
for i, num in enumerate(total):
    plt.text(i, num-0.03, f'{num}', ha='center', va='bottom', fontsize=14, color='w', weight='extra bold')
    
plt.ylim(0, 1.5)  # 첫 번째 y 축 한계 설정

plt.text(5.5, bank[-1], '은행', ha='center', va='bottom', fontsize=14, color='black', weight='extra bold')
plt.text(5.5, non_bank[-1], '비은행', ha='center', va='bottom', fontsize=14, color='black', weight='extra bold')
plt.text(5.5, total[-1], '전체', ha='center', va='bottom', fontsize=14, color='black', weight='extra bold')


# 그래프 제목 설정
plt.title('< 임금근로자 연체율(대출잔액 기준) >', fontsize=20, weight='extra bold')


# 그리드, y 축, 프레임 없애기
plt.tick_params(axis='x', labelsize=15)
plt.grid(False)
plt.yticks([])
plt.gca().spines['left'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['bottom'].set_visible(True)

# 범례 없애기
plt.legend().set_visible(False)

# x 축 라벨 설정
plt.xlabel('', fontsize=14)

# 그래프 표시
plt.show()
```
