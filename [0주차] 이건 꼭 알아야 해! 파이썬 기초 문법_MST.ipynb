## 데이터 추출하기 ─ 슬라이싱 구문, range 메서드
### 1. 슬라이싱 구문으로 데이터 추출하기

```python
subset = df.loc[:, ['법정동', '거래금액']] 
print(subset.head())
```

```python
subset = df.iloc[:, [2, 11, -17]] 
print(subset.head())
```

### 2. range 메서드로 원하는 데이터 추출하기

```python
small_range = list(range(10)) 
print(small_range)
```

```python
print(type(small_range))
```

```python
subset = df.iloc[:, small_range] 
print(subset.head())
```

```python
small_range = list(range(2, 6)) 
print(small_range)
```

```python
subset = df.iloc[:, small_range] 
print(subset.head())
```

```python
small_range = list(range(0, 6, 2)) #0에서 6까지 2씩 띄어가며. 0,2,4 # 마지막 숫자 (6)는 포함되지 않음.
subset = df.iloc[:, small_range] 
print(subset.head())
```

### 4. 슬라이싱, range 메서드 비교하기

```python
subset = df.iloc[:, :3] # 전체 행에 대해서 3번 컬럼 전까지 (0,1,2 컬럼)
print(subset.head())
```

```python
subset = df.iloc[:, 0:6:2] 
print(subset.head())
```

# 연습문제(Exercise) - Just Do It Now !
## 그룹화한 데이터의 평균 구하기_ groupby 함수
### 1. '거래금액' 열을 연도별로 그룹화하여 평균 계산하기

```python
# print(df.head(n=10))
```

```python
print(df.groupby('법정동')['거래금액'].mean())
```

```python
grouped_dong_df = df.groupby('법정동')
print(type(grouped_dong_df))
```

```python
print(grouped_dong_df)
```

```python
grouped_dong_df_price = grouped_dong_df['거래금액'] 
print(type(grouped_dong_df_price))
```

```python
mean_price_by_dong = grouped_dong_df_price.mean() 
print(mean_price_by_dong)
```

### 6. 거래금액 열의 평균값을 법정동, 전용면적으로 그룹화하여 한 번에 계산하기

```python
multi_group_var = df.groupby(['법정동', '전용면적'])[['거래금액']].mean() 
print(multi_group_var)
```

```python
print(type(multi_group_var))
```

### 7. 그룹화한 데이터의 개수 세기

```python
print(df.groupby('법정동')['거래금액'].nunique())
```
